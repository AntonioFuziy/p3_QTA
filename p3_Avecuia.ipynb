#### Reading Data

```python
dbo = pd.read_excel("dbo.xlsx", engine='openpyxl')
turbidez = pd.read_excel("turbidez.xlsx", engine='openpyxl')
nitrogenio = pd.read_excel("nitrogenio.xlsx", engine='openpyxl')
oxigenio = pd.read_excel("oxigenio.xlsx", engine='openpyxl')
ph = pd.read_excel("ph.xlsx", engine='openpyxl')
temperatura = pd.read_excel("temperatura.xlsx", engine='openpyxl')
fosforo = pd.read_excel("fosforo.xlsx", engine='openpyxl')
residuos = pd.read_excel("residuos.xlsx", engine='openpyxl')
```
