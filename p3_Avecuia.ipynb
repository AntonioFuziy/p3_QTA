**Reading all the csv data collected from the IQA Graphics given**

```python
turb_csv = pd.read_csv("./csv/turbidez.csv")
nit_csv = pd.read_csv("./csv/nit_total.csv")
dbo_csv = pd.read_csv("./csv/dbo.csv")
fosf_csv = pd.read_csv("./csv/fosforo.csv")
temp_csv = pd.read_csv("./csv/temperatura.csv")
res_csv = pd.read_csv("./csv/residuos.csv")
ph_csv = pd.read_csv("./csv/ph.csv")
coli_csv = pd.read_csv("./csv/coliformes.csv")
oxi_csv = pd.read_csv("./csv/oxi_porcentagem.csv")
```

```python
q_data = pd.DataFrame()
q_data["DBO"] = dbo_csv.iloc[0:,1]
q_data["Nitrogenio"] = nit_csv.iloc[0:,1]
q_data["Coliformes"] = coli_csv.iloc[:,1]
q_data["pH"] = ph_csv.iloc[:,1]
q_data["Oxigenio"] = oxi_csv.iloc[:,1]
q_data["Temperatura"] = temp_csv.iloc[:,1]
q_data["Fosforo"] = fosf_csv.iloc[:,1]
q_data["Turbidez"] = turb_csv.iloc[:,1]
q_data["Residuos"] = res_csv.iloc[0:,1]
q_data
```

___
**w_list**

```python
w_pH = 0.12
w_DBO = 0.10
w_OD = 0.17
w_temp = 0.10
w_nitro = 0.10
w_turb = 0.08
w_fos = 0.1
w_res = 0.08
w_fosf = 0.1
w_coli = 0.15
w_list = [w_DBO, w_nitro, w_coli, w_pH, w_OD, w_temp, w_fosf, w_turb, w_res]
```

___
**IQA calculate function**

```python
def IQA(q_list, w_list):
    IQA_value = 1
    for i in range(0, len(q_list)):
        IQA_value *= q_list[i]**w_list[i]
    return IQA_value
```

```python
q1 = q_data.loc[0]
q2 = q_data.loc[1]
q3 = q_data.loc[2]
q4 = q_data.loc[3]
q5 = q_data.loc[4]
q6 = q_data.loc[5]
q7 = q_data.loc[6]
q8 = q_data.loc[7]
```

```python
iqa1 = IQA(q1, w_list)
iqa2 = IQA(q2, w_list)
iqa3 = IQA(q3, w_list)
iqa4 = IQA(q4, w_list)
iqa5 = IQA(q5, w_list)
iqa6 = IQA(q6, w_list)
iqa7 = IQA(q7, w_list)
iqa8 = IQA(q8, w_list)
```

```python
iqa_list = [iqa1, iqa2, iqa3, iqa4, iqa5, iqa6, iqa7, iqa8]
```

```python
iqa_dates = [_data.index[1], _data.index[4], _data.index[7], _data.index[10], _data.index[13], _data.index[16], _data.index[18], _data.index[19]]
print(iqa_dates)
```
