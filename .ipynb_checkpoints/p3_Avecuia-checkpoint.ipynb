```python
dbo = pd.read_excel("dbo.xlsx", engine='openpyxl')
turbidez = pd.read_excel("turbidez.xlsx", engine='openpyxl')
nitrogenio = pd.read_excel("nitrogenio.xlsx", engine='openpyxl')
oxigenio = pd.read_excel("oxigenio.xlsx", engine='openpyxl')
ph = pd.read_excel("ph.xlsx", engine='openpyxl')
temperatura = pd.read_excel("temperatura.xlsx", engine='openpyxl')
fosforo = pd.read_excel("fosforo.xlsx", engine='openpyxl')
residuos = pd.read_excel("residuos.xlsx", engine='openpyxl')
ecolif = pd.read_excel("ecoliformes.xlsx", engine='openpyxl')
```

```python
coliformes = pd.DataFrame()
coliformes["Valor"] = ecolif["Valor"]*1.25
```

```python
_data = pd.DataFrame()
_data["DBO"] = dbo["Valor"]
_data["Oxigenio"] = oxigenio["Valor"]
_data["Nitrogenio"] = nitrogenio["Valor"]
_data["Fosforo"] = fosforo["Valor"]
_data["Temperatura"] = temperatura["Valor"]
_data["Residuos"] = residuos["Valor"]
_data["pH"] = ph["Valor"]
_data["Turbidez"] = turbidez["Valor"]
_data["Coliformes"] = coliformes["Valor"]
_data = _data.set_index(pd.to_datetime(dbo["Data Coleta"].values))
_data
```
